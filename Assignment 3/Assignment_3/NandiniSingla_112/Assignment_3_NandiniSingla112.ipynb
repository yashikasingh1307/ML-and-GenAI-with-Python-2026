```python
sum = 0
def SumOfNum(n):
  sum = 0
  for i in range(1,n+1):
    sum += i
  print(sum)
SumOfNum(8)
```

```python
def reverse_num(n):
    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
    return reversed_num

reverse_num(6397529)
```

```python
def count(n):
    n = abs(n)
    if n == 0:
        return 1

    count = 0
    while n > 0:
        count += 1
        n //= 10
    return count
count(6452378)
```

```python
def func(n):
  dig = []
  while n > 0:
    d = n % 10
    dig.append(d)
    n = n//10
  reverse = dig[::-1]
  if dig == reverse:
    print("yes it is a palindrome")
  else:
    print("no it is not a palindrome")


num = int(input("enter your number: "))
func(num)
```

```python
def fibonacci_series(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

n = int(input("Enter the number of terms: "))
fibonacci_series(n)
```
