```python
#Find simple interest
principal = float(input("Enter principal:"))
rate = float(input("Enter rate per year:"))
time = float(input("Enter time in year:"))

interest =(principal*rate*time)/100

print("simple interest =", interest)

```

```python
#7. Student Result System
name = input("Enter name = ")
roll_no = int(input("Enter roll no = "))
subjects = int(input("Enter number of subjects = "))
total = 0
for i in range (0, subjects):
  marks = float(input(f"Enter marks in subject{i}: "))
  total +=marks
score = (total/subjects)

if (score>90):
  print("A+ grade")
elif (score>80):
  print("A grade")
elif(score>70):
  print("B+ grade")
elif (score>60):
  print("B grade")
else:
  print("Fail")
```
