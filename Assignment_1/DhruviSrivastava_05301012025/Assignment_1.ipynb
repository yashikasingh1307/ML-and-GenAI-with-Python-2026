Q1 Find area of rectangle.

```python
length = 20
breadth = 10
area = length * breadth
print("Area of rectangle =", area)
```

Q2 Find simple interest.

```python
principal = 300
rate = 5
time = 3
simple_interest = (principal * rate * time) / 100
print("Simple Interest =", simple_interest)
```

Q3 Convert temperature from Celsius to Fahrenheit.

```python
celsius = 37
fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit =", fahrenheit)
```

Q4 Calculate average of 3 numbers.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))
average = (num1 + num2 + num3) / 3
print("Average of three numbers =", average)
```

Q5 Find square and cube of a number.

```python
number = int(input("Enter a number: "))
square = number ** 2
cube = number ** 3
print("Square =", square)
print("Cube =", cube)
```

Q6 Swap two numbers without third variable.

```python
a = int(input("Enter first number (a): "))
b = int(input("Enter second number (b): "))
print("Before swapping: a =", a, " b =", b)
a = a + b
b = a - b
a = a - b
print("After swapping: a =", a, " b =", b)
```

Q7 Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation

```python
# Taking student details
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")

# Storing marks in variables
english = float(input("Enter marks in English: "))
math = float(input("Enter marks in Math: "))
science = float(input("Enter marks in Science: "))

# Calculating total and percentage
total = english + math + science
percentage = (total / 300) * 100   # Assuming each subject is out of 100

# Displaying student report
print("\n--- Student Report ---")
print("Name:", name)
print("Roll Number:", roll_no)
print("English:", english)
print("Math:", math)
print("Science:", science)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
