```python
# Find sum of first 10 natural number
sum = 0
for i in range (1,11):
  sum+=i
print("Sum of first 10 natural number : ",sum)
```

```python
# Find factorial of a number
num = int(input("Enter a number : "))
fact = 1
for i in range(1,num+1):
  fact*=i
print("Factorial of ",num," is ",fact)
```

```python
# Print Fibonacci series
num = int(input("Enter a number : "))
a,b = 0, 1
for i in range (1, num+1):
  print (a, ' ')
  a, b = b, a+b
```

```python
# Find largest among 3 numbers
num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))
num3 = int(input("Enter third number : "))
if num1>num2 and num1>num3:
  print(num1," is largest")
elif num2>num1 and num2>num3:
  print(num2," is largest")
else:
  print(num3," is largest")
```

```python
from typing import TextIO
# Student Result System

#INPUT STUDENT DETAILS
roll = int(input("Enter roll number : "))
name = input("Enter name : ")

#INPUT MARKS OBTAINED IN SUBJECTS
num_subjects = int(input("Enter number of subjects : "))
marks = []
for i in range(1,num_subjects+1):
  mark = int(input("Enter marks obtained : "))
  marks.append(mark)

#CALCULATE MARKS AND PERCENTAGE
sum = 0
for val in marks :
  sum+=val
total = sum
percentage = total/num_subjects

#OUTPUT DISPLAYED ON SCREEN
print("\nSTUDENT REPORT")
print("\nRoll Number: ", roll)
print("Name: ", name)
print("Total Grade", total)
print("Percentage: ",percentage,"%")

#DISPLAYING GRADE
if percentage>=90:
  print("Grade: A")
elif percentage>=80:
  print("Grade: B")
elif percentage>=70:
  print("Grade: C")
elif percentage>=60:
  print("Grade: D")
else:
  print("Grade: E")

```
