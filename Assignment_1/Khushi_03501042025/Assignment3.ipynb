```python
# Create a function to print first 10 natural numbers.
def first_ten_natural_numbers():
  for i in range(1,11):
    print(i)
first_ten_natural_numbers()
```

```python
# Create a functio to calculate sum of first N natural numbers.
def sum_of_natural_numbers(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum
num = int(input("Enter a number: "))
result = sum_of_natural_numbers(num)
print("The sum of first", num, "natural numbers is", result)
```

```python
# Create a function to reverse a number.
def reverse_number(num):
    rev_num = 0
    while num > 0:
        digit = num % 10
        rev_num = rev_num * 10 + digit
        num //= 10
    return rev_num
reverse_number(145)
```

```python
# Create a function to count digits in a number
def count_digits(num):
    count = 0
    while num > 0:
        num //= 10
        count += 1
    return count
count_digits(12345)
```

```python
# Create a function to check a palindrome number
def check_palindrome(num):
  original_num = num
  rev_num = 0
  while num > 0:
    digit = num % 10
    rev_num = rev_num * 10 + digit
    num //= 10
  if original_num == rev_num:
    return True
  else:
    return False
number = int(input("Enter a number: "))
if check_palindrome(number):
  print(number, "is a palindrome number.")
else:
  print(number, "is not a palindrome number")
```

```python
# Create a ffunction to generate Fibonacci series
def generate_fibonacci(n):
  fib_series = [0, 1]
  for i in range(2, n):
    fib_series.append(fib_series[i-1] + fib_series[i-2])
  return fib_series
generate_fibonacci(10)
```

```python
filename = "student_details.txt"

while True:
    print("\nEnter student details (or type 'exit' for name to quit):")
    student_name = input("Enter student's name: ")

    if student_name.lower() == 'exit':
        break

    student_age = input("Enter student's age: ")
    student_grade = input("Enter student's grade: ")

    # Open the file in append mode ('a') and write the details
    with open(filename, 'a') as file:
        file.write(f"Name: {student_name}\n")
        file.write(f"Age: {student_age}\n")
        file.write(f"Grade: {student_grade}\n")
        file.write("--------------------\n")

    print("Details aadded successfully ",filename)

print("Finished. Thanks for using.")
```
