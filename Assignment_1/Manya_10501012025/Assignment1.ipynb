```python
length = int(input("Enter length of the rectangle: "))
breadth = int(input("Enter breadth of the rectangle: "))
area = length * breadth
print("Area of the rectangle:", area)
```

```python
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest: "))
time = float(input("Enter the time (in years): "))
simple_interest = (principal * rate * time) / 100
print("Simple Interest:", simple_interest)
```

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (9/5 * celsius) + 32
print("Temperature in Fahrenheit =", fahrenheit)
```

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))
average = (a + b + c) / 3
print("Average:", average)
```

```python
num = float(input("Enter a number: "))
square = num ** 2
cube = num ** 3
print("Square:", square)
print("Cube:", cube)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Before swap: a =", a, ", b =", b)
a, b = b, a
print("After swap: a =", a, ", b =", b)
```

```python
# Student Report Program
# Takes student details, calculates total and percentage

# Input student details
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")

# Input marks for 5 subjects
maths = float(input("Enter marks in Maths (out of 100): "))
science = float(input("Enter marks in Science (out of 100): "))
english = float(input("Enter marks in English (out of 100): "))
hindi = float(input("Enter marks in Hindi (out of 100): "))
computer = float(input("Enter marks in Computer (out of 100): "))

# Calculate total and percentage
total = maths + science + english + hindi + computer
percentage = (total / 500) * 100

# Display student report
print("\n========= STUDENT REPORT =========")
print("Name      :", name)
print("Roll No.  :", roll_no)
print("Maths     :", maths)
print("Science   :", science)
print("English   :", english)
print("Hindi     :", hindi)
print("Computer  :", computer)
print("-----------------------------------")
print("Total     :", total, "/ 500")
print("Percentage:", percentage, "%")
print("===================================")
```
