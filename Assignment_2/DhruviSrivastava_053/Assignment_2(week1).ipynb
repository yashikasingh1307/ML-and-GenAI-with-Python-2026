Q1 Find sum of first 10 natural numbers.

```python
total = 0
for i in range(1, 11):
    total += i
print("Sum of first 10 natural numbers:", total)
```

Q2 Find factorial of a number.

```python
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial of", num, "is:", fact)
```

Q3 Print Fibonacci Series.

```python
n = int(input("Enter number of terms: "))
a, b = 0, 1
print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

Q5 Create Student Result System
	-	Input student details
	-	Input marks
	-	Calculate percentage
	-	Display grade
	-	Use:
if-elif-else
loops
user input

```python
name = input("Enter student name: ")
roll = input("Enter roll number: ")

# Input marks
subjects = int(input("Enter number of subjects: "))
total_marks = 0

for i in range(1, subjects + 1):
    marks = float(input(f"Enter marks for subject {i}: "))
    total_marks += marks

# Calculate percentage
percentage = total_marks / subjects

# Display grade using if-elif-else
if percentage >= 90:
    grade = "A+"
elif percentage >= 75:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "F"

# Display result
print("\n--- Student Result ---")
print("Name:", name)
print("Roll Number:", roll)
print("Total Marks:", total_marks)
print("Percentage:", percentage, "%")
print("Grade:", grade)
```
