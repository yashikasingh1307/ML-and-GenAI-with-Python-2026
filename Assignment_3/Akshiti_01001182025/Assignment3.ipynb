```python
#function to print first 10 Natural Numbers
def print_naturalnumbers():
    for i in range(1, 11):
        print(i)

print_naturalnumbers()
```

```python
#Function to Calculate Sum of First N Natural Numbers

def sum_natural(n):
    return n * (n + 1) // 2

n = 10
print("Sum =", sum_natural(n))
```

```python
#Function to Reverse a Number

def reverse_number(num):
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10

    return rev

num = 12345
print(" The final Reversed Number =", reverse_number(num))
```

```python
#Function to Count Digits in a Number

def count_digits(num):
    count = 0

    while num > 0:
        count += 1
        num = num // 10

    return count

num = 123456
print(" The Number of Digits =", count_digits(num))
```

```python
#Function to Check Palindrome Number

def palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10

    if original == reverse:
        return True
    else:
        return False

num = 121

if palindrome(num):
    print(" It is a Palindrome Number")
else:
    print("It is Not a Palindrome Number")
```

```python
#Function to Generate Fibonacci Series

def fibonacci(n):
    a = 0
    b = 1

    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(10)
```

```python
#Calculator Using Functions

def add(a, b):
    print("Result =", a + b)

def subtract(a, b):
    print("Result =", a - b)

def multiply(a, b):
    print("Result =", a * b)

def divide(a, b):
    print("Result =", a / b)

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Enter your choice: "))

if choice == 1:
    add(num1, num2)

elif choice == 2:
    subtract(num1, num2)

elif choice == 3:
    multiply(num1, num2)

elif choice == 4:
    divide(num1, num2)

else:
    print("Wrong Choice")

```
