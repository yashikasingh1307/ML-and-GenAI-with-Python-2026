```python
# Q1 - Print First 10 Natural Numbers
def print_first_10():
    print("First 10 Natural Numbers:")
    for i in range(1, 11):
        print(i, end=" ")
    print()

print_first_10()
```

```python
# Q2 - Sum of First N Natural Numbers
def sum_of_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

n = int(input("Enter N: "))
result = sum_of_n(n)
print(f"Sum of first {n} natural numbers = {result}")
```

```python
# Q3 - Reverse a Number
def reverse_number(num):
    reversed_num = int(str(abs(num))[::-1])
    if num < 0:
        return -reversed_num
    return reversed_num

num = int(input("Enter a number: "))
print(f"Reversed number = {reverse_number(num)}")
```

```python
# Q4 - Count Digits in a Number
def count_digits(num):
    count = 0
    num = abs(num)
    if num == 0:
        return 1
    while num > 0:
        num //= 10
        count += 1
    return count

num = int(input("Enter a number: "))
print(f"Number of digits in {num} = {count_digits(num)}")
```

```python
# Q5 - Check Palindrome Number

def is_palindrome(num):
    original = str(abs(num))
    reversed_str = original[::-1]
    return original == reversed_str

num = int(input("Enter a number: "))
if is_palindrome(num):
    print(f"{num} is a Palindrome number.")
else:
    print(f"{num} is NOT a Palindrome number.")
```

```python
# Q6 - Generate Fibonacci Series

def fibonacci(n):
    a, b = 0, 1
    print(f"Fibonacci series up to {n} terms:")
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print()

n = int(input("Enter number of terms: "))
fibonacci(n)
```

```python
# Q7 - Calculator Using Functions
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero!"
    return a / b
def calculator():
    print("========== Calculator ==========")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("================================")
    choice = input("Select operation (1/2/3/4): ")
    if choice not in ['1', '2', '3', '4']:
        print("Invalid choice! Please select between 1 and 4.")
        return
    a = float(input("Enter first number : "))
    b = float(input("Enter second number: "))
    if choice == '1':
        print(f"Result: {a} + {b} = {add(a, b)}")
    elif choice == '2':
        print(f"Result: {a} - {b} = {subtract(a, b)}")
    elif choice == '3':
        print(f"Result: {a} * {b} = {multiply(a, b)}")
    elif choice == '4':
        print(f"Result: {a} / {b} = {divide(a, b)}")
calculator()
```

```python
# Q8 - Create a Text File and Store Student Details
def store_student_details():
    filename = "students.txt"
    n = int(input("How many students do you want to add? "))

    with open(filename, "w") as f:
        f.write("Name, Marks\n")
        f.write("-" * 20 + "\n")
        for i in range(n):
            print(f"\nEnter details for Student {i + 1}:")
            name  = input("  Name  : ")
            marks = input("  Marks : ")
            f.write(f"{name}, {marks}\n")

    print(f"\nStudent details saved successfully to '{filename}'!")

store_student_details()
```

```python
# Q9 - Read Data from a File

def read_student_details():
    filename = "students.txt"
    try:
        with open(filename, "r") as f:
            print(f"Reading data from '{filename}':\n")
            print(f.read())
    except FileNotFoundError:
        print(f"Error: '{filename}' not found!")
        print("Please run q8_write_file.py first to create the file.")

read_student_details()
```

```python
# Q10 - Handle Division by Zero Using Exception Handling

def safe_divide():
    try:
        a = float(input("Enter numerator   : "))
        b = float(input("Enter denominator : "))
        result = a / b
        print(f"Result: {a} / {b} = {result}")
    except ZeroDivisionError:
        print("Exception caught: Division by zero is not allowed!")
    except ValueError:
        print("Exception caught: Please enter valid numbers!")

safe_divide()
```

```python
# Q11 - Student Class with Name and Marks

class Student:
    def __init__(self, name, marks):
        self.name  = name
        self.marks = marks

    def display(self):
        print(f"Student Name : {self.name}")
        print(f"Marks        : {self.marks}")
        print(f"Grade        : {self.get_grade()}")

    def get_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        elif self.marks >= 40:
            return "D"
        else:
            return "F"


name  = input("Enter student name  : ")
marks = int(input("Enter student marks : "))

s = Student(name, marks)
print()
s.display()
```
