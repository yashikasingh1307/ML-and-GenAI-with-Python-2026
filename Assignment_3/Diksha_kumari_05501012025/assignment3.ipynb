```python
#1. Create a function to print first 10 natural numbers.
def ten_natural():
  print("First 10 natural numbers are:")
  for i in range(1,11):
    print(i,end=" ")
ten_natural()
```

```python
#2. Create a function to  calculate sum of first N natural numbers.
def Sum_natural(num):
  summ=0
  for i in range(1,num+1):
    summ+=i
  print("Sum of first",num,"natural numbers is:",summ)
num= int(input("Enter the number of natural numbers you want to add:"))
Sum_natural(num)
```

```python
#3. Create a function to reverse a number.
def num_reverse(num):
  rev= 0
  while num>0:
    num1= num%10
    rev= rev*10+num1
    num= num//10
  print("Reverse of the number is:",rev)
num= int(input("Enter the number you want to reverse:"))
num_reverse(num)
```

```python
#4. Create a function to count digits in a number.
def digit_count(num):
  a= str(num)
  b= list(a)
  length= len(b)
  print("Digits in",num,"are:",length)
num= int(input("Enter the number:"))
digit_count(num)
```

```python
#5. Create a function to check palindrome number.
def palindrome(num):
  rev= 0
  num2= num
  while num2>0:
    num1= num2%10
    rev= rev*10+num1
    num2= num2//10
  if rev==num:
    print(num, "is a palindrome")
  else:
    print(num,"is not a palindrome")
num= int(input("Enter the number:"))
palindrome(num)
```

```python
#6. Create a function to generate Fibonacci series
def Fibonnaci(num):
  series= [1,1]
  for i in range(num):
    series.append(series[-1]+series[-2])
  print("Fibonacci sequence till",num,'terms:',series)
num= int(input("Enter the number of terms of the sequence:"))
Fibonnaci(num)
```
