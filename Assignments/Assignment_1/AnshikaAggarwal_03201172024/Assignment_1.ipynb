```python
# Find the area of rectangle.
length = float(input("Enter length: "))
width = float(input("Enter width: "))

area = length * width
print("Area of rectangle:", area)
```

```python
# Simple interest
principal = float(input("Enter principal: "))
rate = float(input("Enter rate: "))
time = float(input("Enter time: "))

si = (principal * rate * time) / 100
print("Simple interest:", si)
```

```python
# celsius to fahrenheit
celsius = float(input("Enter celsius: "))

fahrenheit = (celsius * 9/5) + 32
print("Fahrenheit:", fahrenheit)
```

```python
#avg of 3 numbers
num1 = float(input("Enter number 1: "))
num2 = float(input("Enter number 2: "))
num3 = float(input("Enter number 3: "))

avg = (num1 + num2 + num3) / 3
print("Average:", avg)
```

```python
#square and cube of a number
num = float(input("Enter number: "))

square = num ** 2
cube = num ** 3
print("Square:", square)
print("Cube:", cube)
```

```python
# swap 2 numbers without third variable
num1 = float(input("Enter number 1: "))
num2 = float(input("Enter number 2: "))

num1, num2 = num2, num1
print("Number 1:", num1)
print("Number 2:", num2)
```

```python
# Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation
name = input("Enter name: ")
roll_no = input("Enter roll number: ")

m1 = float(input("Enter marks 1: "))
m2 = float(input("Enter marks 2: "))
m3 = float(input("Enter marks 3: "))

total = m1+m2+m3
percentage = (total/300)*100

print("Name:",name)
print("Roll No:", roll_no)
print("Total:",total)
print("Percentage:",percentage)
```
