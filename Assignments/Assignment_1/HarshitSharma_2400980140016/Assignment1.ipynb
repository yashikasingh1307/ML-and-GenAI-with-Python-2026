#### Area of rectangle

```python
length=int(input("Enter the length of rectangle(in cm):"))
breadth=int(input("Enter the breadth of the rectangle(in cm):"))
area=2*length*breadth
print(f"Area of the rectangle={area} cm^2")
```

#### Simple Interest

```python
p=int(input("Enter the amount:"))
r=int(input("Enter the rate:"))
t=int(input("Enter the time(in years):"))
si=p*r*t/100
print(f"Simple Interest for amount {p} for rate {r}% in time {t} years would be : {si}")
print(f"Total amount(Principle + Interest) = {si+p}")
```

#### Convert Tempereature from Celsius to Fahrenheit

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit:", fahrenheit)
```

#### Swap two numbers without using third variable

```python
num1=int(input("Enter the num1:"))
num2=int(input("Enter the num2:"))
print(f"Number 1= {num1} and Number 2 = {num2}")
temp=num1+num2
num1=temp-num1
num2=temp-num2
print(f"Number 1= {num1} and Number 2 = {num2}")
```

#### Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation

```python
name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")

math = int(input("Enter marks in Mathematics: "))
science = int(input("Enter marks in Science: "))
english = int(input("Enter marks in English: "))
history = int(input("Enter marks in History: "))
computer = int(input("Enter marks in Computer Science: "))

total = math + science + english + history + computer

percentage = total / 5

print("\n--- Student Report ---")
print("Name:", name)
print("Roll Number:", roll_no)
print("Mathematics:", math)
print("Science:", science)
print("English:", english)
print("History:", history)
print("Computer Science:", computer)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
