```python
#1 Find area of rectangle.
length = float(input("Enter the length of rectangle: "))
breadth = float(input("Enter the breadth of rectangle: "))
print(f"Area of rectangle: {length*breadth} sq.units")
```

```python
#2 Find simple interest.
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest(in %): "))
time = float(input("Enter the time (in years): "))
simple_interest = (principal*rate*time)/100
print(f"SIMPLE INTEREST : {simple_interest}")

```

```python
#3 Convert temperature from Celsius to Fahrenheit.
c = float(input("Enter temperature (in °C): "))
f = (c*1.8)+32
print(f"Temperature in Fahrenheit: {f}F")
```

```python
#4 Calculate average of 3 numbers.
n1 = float(input("Enter 1st number: "))
n2 = float(input("Enter 2nd number: "))
n3 = float(input("Enter 3rd number: "))
avg = (n1+n2+n3)/3
print(f"Average of 3 numbers = {avg}")
```

```python
#5 Find square and cube of a number.
num = float(input("Enter the number: "))
print(f"Square of {num}: {num**2}")
print(f"Cube of {num}: {num**3}")
```

```python
#6 Swap two numbers without third variable.
n1 = float(input("Enter 1st number: "))
n2 = float(input("Enter 2nd number: "))
n1,n2 = n2,n1
print("After Swap - ")
print(f"First number: {n1}")
print(f"Second number: {n2}")
```

```python
#7 Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation

# Taking student details as input
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")

# Taking marks of 3 subjects
marks1 = float(input("Enter marks in Subject 1: "))
marks2 = float(input("Enter marks in Subject 2: "))
marks3 = float(input("Enter marks in Subject 3: "))

# Calculating total marks
total = marks1 + marks2 + marks3

# Calculating percentage
percentage = round((total / 300) * 100,2)

# Displaying the report
print("STUDENT REPORT")
print("Name:", name)
print("Roll Number:", roll_no)
print("Marks in Subject 1:", marks1)
print("Marks in Subject 2:", marks2)
print("Marks in Subject 3:", marks3)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
