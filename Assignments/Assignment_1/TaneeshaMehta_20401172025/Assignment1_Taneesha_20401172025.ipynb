```python
#Q1 find area of rectangle
len = int(input("enter length: "))
width = int(input("enter width: "))
area = len * width
print( "area of the rectangle = " ,area ,"sq units")
```

```python
#Q2 find simple interest
p = int(input("principal = "))
roi = int(input("rate of interest(%) = "))
t = int(input("time(years) = "))
simpleinterest = (p*roi*t)/100
print("simple interest = " ,simpleinterest)
```

```python
#Q3 convert temperature from celsius to fahrenheit
t = int(input(("enter temperture (celcius) : ")))
f = (t*9)/5 + 32
print("temperture in fahrenheit = " ,f)
```

```python
#Q4 calculate average of three numbers
num1 = int(input(" enter 1st number = "))
num2 = int(input(" enter 2nd number = "))
num3 = int(input(" enter 3rd number = "))
avg = (num1+num2+num3)/3
print("average of the three numbers = " ,avg)
```

```python
#Q5 find square and cube of a number
num = int(input("enter number : "))
sq = num**2
cube = num**3
print("square = ", sq )
print("cube = ", cube)
```

```python
#Q6 swap two numbers without a third variable
a = int(input(" enter 1st number = "))
b = int(input(" enter 2nd number = "))
a,b = b,a
print("number 1 =" ,a)
print("number 2 = ",b)
```

```python
#Q7 create student report program that takes student details using input(), stores marks and calculates total and percentage
name = input("enter student name: ")
roll = input("enter roll no: ")
a= int(input("enter marks of sub 1: "))
b= int(input("enter marks of sub 2: "))
c= int(input("enter marks of sub 3: "))
d= int(input("enter marks of sub 4: "))
total = a+b+c+d
percentage = total / 4
print("Name:", name)
print("Roll Number:", roll)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
