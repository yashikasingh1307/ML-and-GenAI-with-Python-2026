```python
# Function to print first 10 natural numbers

def print_numbers():
    for i in range(1, 11):
        print(i)

# Function call
print_numbers()
```

```python
# Function to check palindrome number

def is_palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10

    if original == reverse:
        return True
    else:
        return False

# Input from user
num = int(input("Enter a number: "))

# Function call
if is_palindrome(num):
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
# Function to calculate sum of first N natural numbers

def sum_natural_numbers(n):
    total = 0

    for i in range(1, n + 1):
        total += i

    return total

# Input from user
n = int(input("Enter a number: "))

# Function call
result = sum_natural_numbers(n)

print("Sum of first", n, "natural numbers =", result)
```

```python
# Function to reverse a number

def reverse_number(num):
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10

    return reverse

# Input from user
num = int(input("Enter a number: "))

# Function call
result = reverse_number(num)

print("Reversed Number =", result)
```

```python
# Function to count digits in a number

def count_digits(num):
    count = 0

    while num > 0:
        num = num // 10
        count += 1

    return count

# Input from user
num = int(input("Enter a number: "))

# Function call
result = count_digits(num)

print("Number of digits =", result)
```
