#### Sum of first 10 natural numbers

```python
sum=0
for i in range(1,11):
    sum+=i
print(f"Sum of first 10 natural numbers :{sum}")
```

#### Factorial of a number

```python
num=int(input("Enter the number for which you want to find factorial :"))
fact=1
for i in range(1,num+1):
    fact*=i
print(f"Factorial for number {num} is {fact}")
    
```

#### Fibonnaci Series

```python
num=int(input("Enter the number of term in fabonacci :"))
a,b=0,1
for i in range(1,num+1):
    print(a, end=" ")
    a, b = b, a + b 
```

#### Largest Amoung 3 numbers

```python
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
num3=int(input("Enter third number:"))
if num1>num2:
    if num1>num3:
        print(f"{num1} is largest.")
    else:
        print(f"{num3} is largest.")
else:
    if num2>num3:
        print(f"{num2} is largest.")
    else:
        print(f"{num3} is largest.")
```

#### Create Student Result System

```python
name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")

math = int(input("Enter marks in Mathematics: "))
science = int(input("Enter marks in Science: "))
english = int(input("Enter marks in English: "))
history = int(input("Enter marks in History: "))
computer = int(input("Enter marks in Computer Science: "))

total = math + science + english + history + computer

percentage = total / 5

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

print("\n--- Student Report ---")
print("Name:", name)
print("Roll Number:", roll_no)
print("Mathematics:", math)
print("Science:", science)
print("English:", english)
print("History:", history)
print("Computer Science:", computer)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
print("Grade:", grade)
```
