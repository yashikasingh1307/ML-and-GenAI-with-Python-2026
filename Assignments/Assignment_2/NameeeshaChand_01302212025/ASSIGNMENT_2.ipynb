```python
# Find sum of first 10 natural numbers.
sum=0
num=1
for i in range(10):
  sum+=num
  num=num+1
print("Sum of first 10 natural numbers: ", sum)
```

```python
# Find factorial of a number.
number=int(input("Enter nuber: "))
num=number
fact=1
while(num>0):
  fact=fact*num
  num=num-1
print("Factorial of ",number," is ",fact)
```

```python
# Print Fibonacci Series.
n=int(input("Enter number: "))
print("Fibonacci Series :")
a=0
b=1
if(n==1):
  print(a)
if(n==2):
  print(a)
  print(b)
else:
  print(a)
  print(b)
  for i in range(n-2):
    c=a+b
    print(c)
    a=b
    b=c
```

```python
# Find largest among 3 numbers
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
c=int(input("Enter third number: "))
if(a>b and a>c):
  print(a," is largest")
elif(b>a and b>c):
  print(b," is largest")
else:
  print(c," is largest")
```

```python
# Create Student Result System
#Input student details ----
#Input marks
#Calculate percentage
#Display grade
#Use: – if-elif-else – loops – user input
name=input("Enter name: ")
roll=int(input("Enter roll number: "))
english=int(input("Enter marks in English: "))
maths=int(input("Enter marks in Maths: "))
science=int(input("Enter marks in Science: "))
computer=int(input("Enter marks in Computer: "))
total=english+maths+science+computer
percentage=(total/400)*100
if(percentage>=90):
  grade="A"
elif(percentage>=80):
  grade="B"
elif(percentage>=70):
  grade="C"
elif(percentage>=60):
  grade="D"
else:
  grade="E"
print("Name: ",name)
print("Roll number: ",roll)
print("Total marks: ",total)
print("Percentage: ",percentage)
print("Grade: ",grade)
```
