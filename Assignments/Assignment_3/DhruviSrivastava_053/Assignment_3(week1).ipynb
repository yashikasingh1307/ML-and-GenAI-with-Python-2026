Q1 Create a function to print first 10 natural numbers.

```python
def print_natural_numbers():
    for i in range(1, 11):
        print(i, end=" ")
    print()

print("Natural numbers from 1 to 10:")
print_natural_numbers()
```

Q2 Create a function to calculate sum of first N natural numbers.

```python
def sum_natural_numbers(n):
    return n * (n + 1) // 2

n = int(input("Enter a number: "))
print("Sum of first", n, "natural numbers:", sum_natural_numbers(n))
```

Q3 Create a function to reverse a number.

```python
def reverse_number(num):
    rev = 0
    while num > 0:
        rev = rev * 10 + num % 10
        num //= 10
    return rev

num = int(input("Enter a number: "))
print("Reversed number:", reverse_number(num))
```

Q4 Create a function to count digits in a number.

```python
def count_digits(num):
    return len(str(num))

num = int(input("Enter a number: "))
print("Number of digits:", count_digits(num))
```

Q5 Create a function to check palindrome number.

```python
def is_palindrome(num):
    return str(num) == str(num)[::-1]

num = int(input("Enter a number: "))
if is_palindrome(num):
    print(num, "is a palindrome number.")
else:
    print(num, "is not a palindrome number.")
```

Q6 Create a function to generate Fibonacci series.

```python
def fibonacci_series(n):
    a, b = 0, 1
    series = []
    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    return series

n = int(input("Enter the number of terms: "))
print("Fibonacci series:", fibonacci_series(n))
```

Q7 Calculator Using Functions that contains the following features;
	-	User selects operation
	-	Program performs calculation
	-	Display result

```python
def add(x, y): return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y):
    try:
        return x / y
    except ZeroDivisionError:
        return "Error: Division by zero!"

def calculator():
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    choice = input("Enter choice (1/2/3/4): ")
    x = float(input("Enter first number: "))
    y = float(input("Enter second number: "))

    if choice == '1':
        print("Result:", add(x, y))
    elif choice == '2':
        print("Result:", subtract(x, y))
    elif choice == '3':
        print("Result:", multiply(x, y))
    elif choice == '4':
        print("Result:", divide(x, y))
    else:
        print("Invalid choice")

calculator()
```

Q8 Create a text file and store student details. - Read data from the file.

```python

name = input("Enter student name: ")
roll = input("Enter roll number: ")
marks = input("Enter marks: ")

with open("student.txt", "w") as f:
    f.write(f"Name: {name}\n")
    f.write(f"Roll Number: {roll}\n")
    f.write(f"Marks: {marks}\n")

print("\nStudent details stored in student.txt")

try:
    with open("student.txt", "r") as f:
        print("\n--- Student Details from File ---")
        print(f.read())
except FileNotFoundError:
    print("File not found. Please run the program again.")
```

Q10 Create a Student class with name and marks

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

# Example usage
s1 = Student("DHRUVI", 85)
s1.display()
```
