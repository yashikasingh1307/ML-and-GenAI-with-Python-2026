```python
#Function to print first 10 natural numbers
def firstTen():
    for i in range(1,11):
        print(i,end=" ")
if __name__=="__main__":
    firstTen()
```

```python
#function to calculate sum of first 10 natural numbers
def sumFirstTen():
    sum=0
    for i in range(1,11):
        sum+=i
    return sum
if __name__=="__main__":
    print(sumFirstTen())
```

```python
#function to reverse a number
def reverse(num):
    reverseNum=0
    while num>0:
        digit=num%10
        reverseNum=(reverseNum*10)+digit
        num//=10
    return reverseNum
if __name__=="__main__":
    try:
        num=int(input("Enter the number:"))
        result=reverse(num)
        print(f"Number is {num} and Reverse number is {result}")
    except ValueError:
        print("Enter the numeric Value")
```

```python
#function to count digits in the number
def countDigits(num):
    count=0
    while num>0:
        count+=1
        num//=10
    return count
if __name__=="__main__":
    try:
        num=int(input("Enter the number:"))
        result=countDigits(num)
        print(f"Number is {num} and Digits are {result}")
    except ValueError:
        print("Enter the numeric Value")
```

```python
#function to check palindrome number
def checkPalindrome(num):
    reverseNum=int(str(num)[::-1])
    if num==reverseNum:
        return True
    else:
        return False
if __name__=="__main__":
    try:
        num=int(input("Enter the num:"))
        print(f"Num is {num} and Is it palindrome : {checkPalindrome(num)} ")
    except ValueError:
        print("Enter numeric Values of num")
```

```python
#function to generate fibonacci series
def fibo(num):
    a,b=0,1
    for _ in range(1,num+1):
        print(a,end=" ")
        a,b=b,a+b

if __name__=="__main__":
    try:
        num=int(input("Enter the num:"))
        fibo(num)
    except ValueError:
        print("Enter the numeric value only.")
```

```python
#calculator using function
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Cannot divide by zero."
    return x / y

def calculator():
    print("Select an operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    choice = input("Enter choice (1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter numerical values.")
            return

        if choice == '1':
            print(f"Result: {add(num1, num2)}")
        elif choice == '2':
            print(f"Result: {subtract(num1, num2)}")
        elif choice == '3':
            print(f"Result: {multiply(num1, num2)}")
        elif choice == '4':
            print(f"Result: {divide(num1, num2)}")
    else:
        print("Invalid selection.")

if __name__=="__main__":
    calculator()
```

```python
#Create a text file and store student details
def write_students_to_file(filename):
    student_details = [
        "ID: 1 | Name: Alice | Grade: A",
        "ID: 2 | Name: Bob   | Grade: B",
        "ID: 3 | Name: Carol | Grade: A+"
    ]

    # "w" mode creates the file if it doesn't exist, or overwrites it if it does
    with open(filename, "w") as file:
        for student in student_details:
            file.write(student + "\n")

    print(f"Student details successfully saved to {filename}")

write_students_to_file("students.txt")
```

```python
#Read data from a file
def read_file_contents(filename):
    try:
        with open(filename, "r") as file:
            data = file.read()
            print(f"--- Contents of {filename} ---")
            print(data)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please run Program 2 first.")

read_file_contents("students.txt")
```

```python
#Handle division by zero using exception handling
def safe_divide(numerator, denominator):
    try:
        result = numerator / denominator
        print(f"{numerator} divided by {denominator} is: {result}")
    except ZeroDivisionError:
        print("Exception Caught: Attempted to divide by zero! Please provide a non-zero denominator.")

# Testing the function
print("Test 1 (Valid):")
safe_divide(10, 2)

print("\nTest 2 (Division by Zero):")
safe_divide(5, 0)
```

```python
#Create a Student class with name and marks
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def display_details(self):
        print(f"Student Name: {self.name}")
        print(f"Marks Secured: {self.marks}")
        print("-" * 25)

# Creating instances (objects) of the Student class
student1 = Student("Janvi", 95)
student2 = Student("Pihu", 78)

# Calling the display method on the objects
print("Student Records:")
student1.display_details()
student2.display_details()
```
