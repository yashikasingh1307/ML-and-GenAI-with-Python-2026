## Q1: Print first 10 natural numbers

```python
# Question 1: Create a function to print first 10 natural numbers.
def print_first_10():
    for i in range(1, 11):
        print(i)

print_first_10()
```

## Q2: Sum of first N natural numbers

```python
# Question 2: Create a function to calculate sum of first N natural numbers.
def sum_of_n(n):
    return n * (n + 1) // 2

print(f"Sum of first 10 natural numbers: {sum_of_n(10)}")
```

## Q3: Reverse a number

```python
# Question 3: Create a function to reverse a number.
def reverse_number(n):
    return int(str(n)[::-1])

print(f"Reverse of 12345: {reverse_number(12345)}")
```

## Q4: Count digits in a number

```python
# Question 4: Create a function to count digits in a number.
def count_digits(n):
    return len(str(abs(n)))

print(f"Number of digits in 98765: {count_digits(98765)}")
```

## Q5: Check palindrome number

```python
# Question 5: Create a function to check palindrome number.
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

print(f"Is 121 a palindrome? {is_palindrome(121)}")
print(f"Is 123 a palindrome? {is_palindrome(123)}")
```

## Q6: Generate Fibonacci series

```python
# Question 6: Create a function to generate Fibonacci series.
def fibonacci(n):
    series = []
    a, b = 0, 1
    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    return series

print(f"First 10 Fibonacci numbers: {fibonacci(10)}")
```

## Q7: Calculator using functions

```python
# Question 7: Calculator using functions.
# User selects operation, program performs calculation, displays result.
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b

def calculator():
    print("Select operation: 1) Add  2) Subtract  3) Multiply  4) Divide")
    choice = input("Enter choice (1/2/3/4): ")
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    ops = {'1': (add, '+'), '2': (subtract, '-'), '3': (multiply, '*'), '4': (divide, '/')}
    if choice in ops:
        func, symbol = ops[choice]
        print(f"Result: {a} {symbol} {b} = {func(a, b)}")
    else:
        print("Invalid choice.")

calculator()
```

## Q10: Handle division by zero using exception handling

```python
# Question 10: Handle division by zero using exception handling.
def safe_divide(a, b):
    try:
        result = a / b
        print(f"{a} / {b} = {result}")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")

safe_divide(10, 2)
safe_divide(10, 0)
```

## Q11: Create a Student class with name and marks

```python
# Question 11: Create a Student class with name and marks.
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Student Name: {self.name}, Marks: {self.marks}")

    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        else:
            return "F"

s1 = Student("Alice", 92)
s2 = Student("Bob", 74)
s1.display()
print(f"Grade: {s1.grade()}")
s2.display()
print(f"Grade: {s2.grade()}")
```
