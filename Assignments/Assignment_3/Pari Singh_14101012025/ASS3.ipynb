```python

# Q1. Create a function to print first 10 natural numbers.


def print_natural_numbers():
    print("First 10 Natural Numbers:")
    for i in range(1, 11):
        print(i, end=" ")
    print("\n")



# Q2. Create a function to calculate sum of first N natural numbers.


def sum_natural_numbers(n):
    return n * (n + 1) // 2



# Q3. Create a function to reverse a number.


def reverse_number(num):
    reverse = 0
    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num //= 10
    return reverse



# Q4. Create a function to count digits in a number.


def count_digits(num):
    if num == 0:
        return 1

    count = 0
    while num > 0:
        count += 1
        num //= 10
    return count



# Q5. Create a function to check palindrome number.


def is_palindrome(num):
    return num == reverse_number(num)


# Q6. Create a function to generate Fibonacci series.


def fibonacci(n):
    a = 0
    b = 1
    print("Fibonacci Series:")
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print("\n")


# Q7. Calculator Using Functions
# Features:
# 1. User selects operation
# 2. Program performs calculation
# 3. Display result


def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Division by zero is not allowed."
    return a / b


# Q8. Create a text file and store student details.


def write_student_file():
    file = open("students.txt", "w")

    name = input("Enter Student Name: ")
    roll = input("Enter Roll Number: ")
    marks = input("Enter Marks: ")

    file.write("Name: " + name + "\n")
    file.write("Roll Number: " + roll + "\n")
    file.write("Marks: " + marks + "\n")

    file.close()
    print("Student details saved successfully.\n")


# Q9. Read data from a file.


def read_student_file():
    try:
        file = open("students.txt", "r")
        print("\nStudent Details:")
        print(file.read())
        file.close()
    except FileNotFoundError:
        print("File not found.\n")


# Q10. Handle division by zero using exception handling.


def exception_demo():
    try:
        num1 = int(input("Enter first number: "))
        num2 = int(input("Enter second number: "))
        result = num1 / num2
        print("Result =", result)
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
    print()


# Q11. Create a Student class with name and marks.


class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)
        print()


# MAIN PROGRAM


# Q1
print_natural_numbers()

# Q2
n = int(input("Enter N to find sum of first N natural numbers: "))
print("Sum =", sum_natural_numbers(n))
print()

# Q3
num = int(input("Enter a number to reverse: "))
print("Reversed Number =", reverse_number(num))
print()

# Q4
num = int(input("Enter a number to count digits: "))
print("Total Digits =", count_digits(num))
print()

# Q5
num = int(input("Enter a number to check palindrome: "))
if is_palindrome(num):
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
print()

# Q6
terms = int(input("Enter number of Fibonacci terms: "))
fibonacci(terms)

# Q7
print("Calculator")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("Enter your choice (1-4): "))
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

if choice == 1:
    print("Result =", add(a, b))
elif choice == 2:
    print("Result =", subtract(a, b))
elif choice == 3:
    print("Result =", multiply(a, b))
elif choice == 4:
    print("Result =", divide(a, b))
else:
    print("Invalid Choice")
print()

# Q8
write_student_file()

# Q9
read_student_file()

# Q10
exception_demo()

# Q11
name = input("Enter Student Name: ")
marks = int(input("Enter Student Marks: "))

student = Student(name, marks)
student.display()
```
