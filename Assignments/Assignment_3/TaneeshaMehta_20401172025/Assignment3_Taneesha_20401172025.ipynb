```python
#Q1 create a function to print first 10 natural numbers
def natural_no():
    print("first 10 natural num")
    for i in range(1,11):
        print(i)
natural_no()
```

```python
#Q2 create a function to calculate sum of first 10 natural numbers
def sum_natural_no(n):
    sum = 0
    for i in range(1,n+1):
        sum += i
    return sum
sum_natural_no(10)
```

```python
#Q3 create a function to reverse a number
def reverse():
    n = int(input("Enter no "))
    a = str(n)
    rev = int(a[ : :-1])
    return rev
print(reverse())
```

```python
#Q4 create a function to count digits in a number
def count_digits(n):
    a = len(str(n))
    return a
count_digits(1234)
```

```python
#Q5 create a function to check palindrome number
def check_palindrome(n):
    a = int(str(n)[: :-1])
    if a == n:
        print(n, "is palindrome")
    else :
        print(n, "is not a palindrome")
print(check_palindrome(1221))
```

```python
#Q6 create a function to check fibonacci series
def fibonacci_series():
    a = 0
    b = 1
    n = int(input("Enter number of fibonacci series"))
    print (a,b, end = " ")
    for i in range(n):
        res = a  + b
        a = b
        b = res
        print(res , end = " ")
fibonacci_series()
```

```python
#Q7 calculator using functions (select operation, perform calculation, display result)
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b == 0:
        return "Division by zero is not allowed"
    return a / b
print("Operations:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
choice = int(input("Enter your choice (1-4): "))
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
if choice == 1:
    result = add(num1, num2)
elif choice == 2:
    result = subtract(num1, num2)
elif choice == 3:
    result = multiply(num1, num2)
elif choice == 4:
    result = divide(num1, num2)
else:
    result = "Invalid Choice"
print("Result =", result)
```
