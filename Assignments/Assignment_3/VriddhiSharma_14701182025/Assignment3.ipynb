```python
# Q2:
n=int(input("Enter the number: "))
def sumOfNaturalNo(n):
  sums=0
  for i in range(1,n+1):
    sums+=i
  return sums

sumOfNaturalNo(n)
```

```python
# Q3:
n=int(input("enter the number"))
def reverse(n):
  reverse=0
  while n>0:
    digit=n%10
    reverse=reverse*10 + digit
    n=n//10
  return reverse
reverse(n)
```

```python
# Q4:
n=int(input("enter the number:"))
def countDigits(n):
    if n==0:
        return 1
    count=0
    while n>0:
        n=n//10
        count+=1
    return count
countDigits(n)
```

```python
# Q5:
n=int(input("enter the numbe:"))
def findPalindrome(n):
    string=str(n)
    palindrome=True
    for i in range(len(string)//2):
        if string[i]!=string[len(string)-i-1]:
            palindrome=False
            break
    return palindrome
findPalindrome(n)
```

```python
# Q6:
n=int(input("enter the number upto which you want fabonacci series:"))
def fabonacci(n):
    if n<=0:
        print("invalid number")
    elif n==1:
        print(1)
    elif n==2:
        print(1,1)
    else:
        prev=1
        prev_prev=1
        print(prev_prev,prev,end=" ")
        for i in range(3,n+1):
            print(prev_prev+prev,end=" ")
            prev_prev,prev=prev,prev_prev+prev
    return

fabonacci(n)
```

```python
# Q7:

def Calculator(a,b,op):
    if op=='+':
        print("result:",a+b)
    elif op=='-':
        print("result:",a-b)
    elif op=='*':
        print("result:",a*b)
    elif op=='/':
        if b!=0:
            print("result:",a/b)
        else:
            print("error:division by zero")
    elif op=='^':
        print("result:",a**b)
    elif op=='%':
        print("result:",a%b)
    elif op=='//':
        print("result:",a//b)
    else:
        print("invalid operator")

while True:
    a=float(input("enter the first number:"))
    op=input("enter the operator:")
    b=float(input("enter the second number:"))

    Calculator(a,b,op)

    choice=input("Do you want to continue?(Yes/No): ")
    if choice.lower()!='yes':
        print("calculator is getting OFF")
        break
```
