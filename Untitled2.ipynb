```python
def print_natural():
    for i in range(1, 11):
        print(i, end=" ")

print_natural()

```

```python
def sum_natural(n):
    return n * (n + 1) // 2

n = int(input("Enter N: "))
print("Sum =", sum_natural(n))
```

```python
def reverse_number(num):
    rev = 0
    while num > 0:
        rev = rev * 10 + num % 10
        num //= 10
    return rev

n = int(input("Enter number: "))
print("Reversed =", reverse_number(n))

```

```python
def count_digits(num):
    count = 0
    while num > 0:
        count += 1
        num //= 10
    return count

n = int(input("Enter number: "))
print("Digits =", count_digits(n))
```

```python
def is_palindrome(num):
    return num == reverse_number(num)

n = int(input("Enter number: "))
if is_palindrome(n):
    print("Palindrome number")
else:
    print("Not a palindrome")
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

n = int(input("Enter number of terms: "))
fibonacci(n)
```

```python
def add(a, b): return a + b
def sub(a, b): return a - b
def mul(a, b): return a * b
def div(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero"

def calculator():
    print("Select operation: +, -, *, /")
    op = input("Enter operation: ")
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    if op == '+':
        print("Result =", add(a, b))
    elif op == '-':
        print("Result =", sub(a, b))
    elif op == '*':
        print("Result =", mul(a, b))
    elif op == '/':
        print("Result =", div(a, b))
    else:
        print("Invalid operation")

calculator()
```

```python
# Create a text file and store student details
with open("students.txt", "w") as file:
    n = int(input("Enter number of students: "))
    for i in range(n):
        name = input(f"Enter name of student {i+1}: ")
        marks = input(f"Enter marks of student {i+1}: ")
        file.write(f"{name},{marks}\n")
print("Student details stored in students.txt")
```

```python
# Read data from a file
with open("students.txt", "r") as file:
    print("Reading student details:")
    for line in file:
        name, marks = line.strip().split(",")
        print(f"Name: {name}, Marks: {marks}")
```

```python
# Handle division by zero using exception handling
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print("Result =", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

```python
# Create a Student class with name and marks
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

# Example usage
s1 = Student("Alice", 85)
s1.display()
```
