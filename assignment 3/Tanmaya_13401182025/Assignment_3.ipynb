```python
#Q2
def natural_numbers(n):
  i=0
  for x in range(0,n+1):
    i+=x
  print(i)
natural_numbers(10)
```

```python
# Q3
def reverse_number(num):
  rev = 0
  while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10
    return rev
print("Reverse =", reverse_number(1234))
```

```python
#Q4
def palindrome(num):
    return num == reverse_number(num)

print("Palindrome =", palindrome(121))
```

```python
#Q5
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

print("\nFibonacci Series:")
fibonacci(10)
```

```python
#Q6
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Enter choice: "))

if choice == 1:
    print("Result =", add(a, b))
elif choice == 2:
    print("Result =", subtract(a, b))
elif choice == 3:
    print("Result =", multiply(a, b))
elif choice == 4:
    print("Result =", divide(a, b))
else:
    print("Invalid Choice")
```
